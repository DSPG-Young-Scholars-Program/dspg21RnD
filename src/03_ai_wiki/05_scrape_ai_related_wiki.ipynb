```python
links_valid = pd.read_csv("/home/zz3hs/git/dspg21RnD/data/dspg21RnD/wiki_ai_links_valid.csv") #import csv
```

```python
links = links_valid["url"]
```

```python
links[0]
```

```python
articles = pd.DataFrame()
for link in links:
    url = "https://en.wikipedia.org/" + link
    print(url)
    source = urllib.request.urlopen(url).read()
    print("Scrapping.")
    soup = BeautifulSoup(source, 'html.parser')
    texts = ""
    for paragraph in soup.select('p'):
        texts += paragraph.text
    articles = articles.append({
        "link": link,
        "url": url,
        'article': texts
        }, ignore_index = True)
```

```python
articles.to_csv(r'/home/zz3hs/git/dspg21RnD/data/dspg21RnD/wiki_ai_related_articles.csv', index = False)   
```
